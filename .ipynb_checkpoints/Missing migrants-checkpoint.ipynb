```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
Migrants = pd.read_csv('./MissingMigrants-Global-2019-12-31_correct.csv', sep = ',')
Migrants
```

```python
Migrants.groupby('Cause of Death').count()[['Location Description']]
```

```python
Migrants = Migrants[Migrants['Cause of Death']=='Unknown']
Migrants.groupby('Number Dead').count()['Cause of Death']
```

```python
Migrants = Migrants[Migrants['Migration Route']=='Central America to US']
Migrants.groupby('Region of Incident').count()['Total Dead and Missing']
```

```python
deaths = Migrants['Number Dead']
avg = np.round(deaths.mean(),2)
avg
```

```python
fig, ax = plt.subplots(figsize=(16, 9)) #Subplots makes a call in the background to create a plot area
# generate random data
data = 'MissingMigrants-Global-2019-12-31_correct.csv'
plt.hist(data, bins = 50, label = 'Data', color = 'r', alpha = 0.75)
plt.xticks(fontsize = 25)
plt.yticks(fontsize = 25)
plt.xlabel('Value', fontname = 'Sans serif', fontsize = 35)
plt.ylabel('Frequency', fontname = 'Sans serif', fontsize = 35)
plt.legend(loc = 'upper left', fontsize = 30)
plt.show()
```
